```
import random
import numpy as np
import matplotlib.pyplot as plt
```

In forgetting model with   consider 2 experimental protocols consisting of 2 trials. In both protocols in first trial find all the words that are remembered out of 500 presented. For second trial move words that are remembered in front of list.  In first protocol, present words in modified order, whereas in second protocol present words in reverse order from the end of list to beginning. Compare the number of words remembered in 2 protocols.

```
T = 500 # Number of elements
n = 5 # dimensions
```

```
# Compare two elements
def compare_elements(elem1,elem2):
  for i in range(len(elem1)):
    if elem1[i] < elem2[i]:
      return True
  return False

# Execute a trial and returns list of remembered and forgotten words
def execute_trial(V):
  remembered = np.array([]) # list of remembered words
  forgotten = [] # list of forgotten words
  for k in range(T):
    for elem in remembered:
      if not compare_elements(V[k], elem):
        forgotten.append(elem)
    remembered = [elem for elem in remembered if compare_elements(V[k], elem)]
    remembered.append(V[k])
  return (remembered, forgotten)
```


### Protocol 1 Trial 1


```
V_11 = np.random.rand(T,n)
(remembered_11, forgotten_11) = execute_trial(V_11)
```


### Protocol 1 Trial 2


```
V_12 = remembered_11 + forgotten_11
(remembered_12, forgotten_12) = execute_trial(V_12)
```

### Protocol 2 Trial 1


```
V_21 = np.random.rand(T,n)
(remembered_21, forgotten_21) = execute_trial(V_21)
```

### Protocol 2 Trial 2


```
V_22 = remembered_21 + forgotten_21
V_22.reverse()

(remembered_22, forgotten_22) = execute_trial(V_22)
```

## Ploting the results

#### Number of remembered words 

```
h=[len(remembered_11),len(remembered_12),len(remembered_21),len(remembered_22)]
plt.plot(["P1T1","P1T2","P2T1","P2T2"],h)
```

We can observe that when we put remembered words on the begining of the list, and that we create a new trial, we have a larger amount of remembered words.

This result is coherent as putting the remembered words at the begining of the list, they will be harder to erase.


If the remembered words are put to the end of the list, they will have more chance to be forgotten
